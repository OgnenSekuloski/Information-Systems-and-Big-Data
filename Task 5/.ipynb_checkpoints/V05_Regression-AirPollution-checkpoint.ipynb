```python
# Create X and y... X contains all the features, y contains the target variable (PM10)
y = pd.DataFrame(data.iloc[:,0])
X = data.drop(['PM10'], axis=1)
```

Standardize the data.

```python
scaler = StandardScaler()
new_data = scaler.fit_transform(X)
col = X.columns #getting the names of the columns
X_standardized = pd.DataFrame(new_data, columns = col) #creating a pandas dataframe with the new data
X_standardized
```

## Split the data 70%-30%

Set the random_state=1

https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_standardized,y,train_size=0.7,random_state=1)
X_train
```

# 3  Regression Learning

## 3.1 Create regressors with default parameters:
- Simple Linear Regression (using only the most correlated feature with the target)
- Linear Regression
- Polynomial with degree 2
- Polynomial with degree 3
- Regression Tree
- Random Forest Regressor
- KNN Regressor
- Dummy regressor

```python
from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures
from sklearn.dummy import DummyRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
```

```python
simple = linear_model.LinearRegression() # we can see that the most correlated is visibility with 0.68
simple.fit(pd.DataFrame(X_train['visibility']), y_train)
y1 = simple.predict(pd.DataFrame(X_test['visibility']))
y1
```

```python
lin = linear_model.LinearRegression()
lin.fit(X_train,y_train)
y2 = lin.predict(X_test)
y2
```

```python
poly2 = PolynomialFeatures(degree=2)
X_poly = poly2.fit_transform(X_train)
X_poly_test = poly2.fit_transform(X_test)
linpoly2 = linear_model.LinearRegression()
linpoly2.fit(X_poly, y_train)
y3 = linpoly2.predict(X_poly_test)
y3
```

```python
poly3 = PolynomialFeatures(degree=3)
X_poly = poly3.fit_transform(X_train)
X_poly_test = poly3.fit_transform(X_test)
linpoly3 = linear_model.LinearRegression()
linpoly3.fit(X_poly, y_train)
y4 = linpoly3.predict(X_poly_test)
y4
```

```python
dt = DecisionTreeRegressor(random_state=1)
dt.fit(X_train,y_train)
y5 = dt.predict(X_test)
y5
```

```python
fr = RandomForestRegressor(n_estimators= 100, random_state=1)
fr.fit(X_train,y_train)
y6 = fr.predict(X_test)
y6
```

```python
knn = KNeighborsRegressor(n_neighbors=3)
knn.fit(X_train,y_train)
y7 = knn.predict(X_test)
y7
```

```python
dummy = DummyRegressor(strategy="mean")
dummy.fit(X_train,y_train)
y8 = dummy.predict(X_test)
y8
```

## 3.2  Compare the MAE, MSE, RMSE and R2 values in a table

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Define models and metrics
models = {
    'Simple Linear Regression': y1,
    'Linear Regression': y2,
    'Polynomial dgr 2': y3,
    'Polynomial dgr 3': y4,
    'Regression Tree': y5,
    'Random Forest Regressor': y6,
    'KNN Regressor': y7,
    'Dummy Regressor': y8,
}

metrics = ['MAE', 'MSE', 'RMSE', 'R2']
results = {}

# metrics for each model
for model_name, y_pred in models.items():
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    results[model_name] = [mae, mse, rmse, r2]

# DataFrame from results
comparison_table = pd.DataFrame(results, index=metrics).T.reset_index()
comparison_table.rename(columns={'index': 'Model'}, inplace=True)

# display format for float values
pd.options.display.float_format = '{:.2f}'.format

comparison_table
```

## 3.3 Visualize the Predictions/Errors for each regressor

Visualize the errors (actual vs predicted)

```python
def make_a_plot(predictions, regressor_name):
    sns.set(style="whitegrid", font_scale=1.2)
    fig, ax = plt.subplots(figsize=(10, 8))
    
    # actual vs predicted values
    ax.scatter(y_test, predictions, color='blue', alpha=0.7, label='Predictions')
    
    # diagonal line for perfect predictions
    ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2, label='Perfect Predictions')
    
    ax.set_xlabel('Actual', fontsize=14)
    ax.set_ylabel('Predicted', fontsize=14)
    ax.set_title('Actual vs Predicted - ' + regressor_name, fontsize=16)
    ax.legend(fontsize=12)
    fig.tight_layout()
    plt.show()
```

```python
make_a_plot(y1, 'Simple Linear Regression')
```

```python
make_a_plot(y3,'Polynomial dgr 2')
```

```python
make_a_plot(y4,'Polynomial dgr 3')
```

```python
make_a_plot(y5,'Regression Tree')
```

```python
make_a_plot(y6,'Random Forest Regressor')
```

```python
make_a_plot(y7,'KNN Regressor')
```

```python
make_a_plot(y8,'Dummy Regressor')
```

# 4. Cross validation

- Perform 5-fold random cross validaiton.
- Use the same algorithms as above.
- Compare the results in a table
- Compare the results for each algorithm between the 70-30 split and the 5 fold CV

```python
from sklearn.model_selection import cross_validate
from sklearn.model_selection import KFold

def evaluate_model(model, X, y, cv):
    scoring = ['neg_mean_absolute_error', 'neg_mean_squared_error', 'neg_root_mean_squared_error', 'r2']
    scores = cross_validate(model, X, y, cv=cv, scoring=scoring)
    return [np.mean(scores['test_neg_mean_absolute_error'] * (-1)),
            np.mean(scores['test_neg_mean_squared_error'] * (-1)),
            np.mean(scores['test_neg_root_mean_squared_error'] * (-1)),
            np.mean(scores['test_r2'])]

models = {
    'Linear Regression': linear_model.LinearRegression(),
    'Polynomial dgr 2': PolynomialFeatures(degree=2),
    'Polynomial dgr 3': PolynomialFeatures(degree=3),
    'Regression Tree': DecisionTreeRegressor(random_state=1),
    'Random Forest Regressor': RandomForestRegressor(n_estimators=100, random_state=1),
    'KNN Regressor': KNeighborsRegressor(n_neighbors=3),
    'Dummy Regressor': DummyRegressor(strategy="mean")
}

scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)

# cross-validation
kf = KFold(n_splits=5, shuffle=True)

results = {}
for model_name, model in models.items():
    if isinstance(model, PolynomialFeatures):
        X_transformed = model.fit_transform(X_standardized)
        results[model_name] = evaluate_model(linear_model.LinearRegression(), X_transformed, y, kf)
    else:
        results[model_name] = evaluate_model(model, X_standardized, y, kf)

# Create comparison table
metrics = ['MAE', 'MSE', 'RMSE', 'R2']
comparison_table1 = pd.DataFrame(results, index=metrics).T.reset_index()
comparison_table1.rename(columns={'index': 'Model'}, inplace=True)

comparison_table1
```
