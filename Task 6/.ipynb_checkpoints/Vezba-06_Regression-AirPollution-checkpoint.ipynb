# 2. Data Preprocessing

## 2.1 NULL values

Remove rows where the PM10 variable is missing or NULL

- use: data.dropna(axis=0, inplace=True, subset=['PM10'])

```python
data.isnull().sum()
```

```python
data.dropna(axis=0, inplace=True, subset=['PM10'])
data
```

use functions do detect how many values are NULL
 - e.g. data.isnull().sum()

```python
data.isnull().sum()
```

Fill the feature values that are missing or NULL.
 - e.g., data['pressure'] = data['pressure'].fillna(data['pressure'].mean())

```python
columns_to_fill = ['windBearing', 'pressure', 'apparentTemperature', 'dewPoint', 'humidity', 'precipIntensity', 'precipProbability', 'temperature', 'visibility', 'windSpeed']

for column in columns_to_fill:
    data[column] = data[column].fillna(data[column].mean())
    
print(data.isnull().sum())
```

### Timestamp... 
- do you need it as feature? 
- can you extract some features from the timestamp? - If yes, extract several (hour, day, month)

* data['hour'] = pd.to_datetime(data['time']).dt.hour

Drop the time feature... it is not needed anymore

```python
data['time'] = pd.to_datetime(data['time'], format='%m/%d/%y %H:%M')

data['hour'] = data['time'].dt.hour
data['day'] = data['time'].dt.day
data['month'] = data['time'].dt.month

data.drop(['time'], axis = 1, inplace= True)
data
```

```python
y = pd.DataFrame(data.iloc[:,1])
X = data.drop(['PM10'], axis=1)
```

## 2.2 Standardize the train and test features. Use StandardScaler
<br><b> do not standardize the target variable!

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_standard = scaler.fit_transform(X)
X_standard
```

## 2.4 Split the data 

Set the random_state=1

```python
#split the data, 70% train and 30% test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_standard, y, train_size = 0.7, random_state=1)
X_train
```

## 3 Default Regressor

Create regressors with default parameters:
- Linear Regression
- LASSO regression
- Random Forest
- SVR
- KNN
- Dummy regressor

Compare the MAE values in a table

```python
from sklearn import linear_model
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.dummy import DummyRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
# LINEAR REGRESSION

simple = linear_model.LinearRegression()

visibility_index = 0  
# Extract 'visibility' column from X_train
visibility_column_train = X_train[:, visibility_index].reshape(-1, 1)

simple.fit(visibility_column_train, y_train)

visibility_column_test = X_test[:, visibility_index].reshape(-1, 1)

y1 = simple.predict(visibility_column_test)
y1
```

```python
# LASSO MODEL

lasso_model = linear_model.Lasso(alpha=0.2)
lasso_model.fit(X_train, y_train)
y2 = lasso_model.predict(X_test)
y2
```

```python
# RANDOM FOREST

fr = RandomForestRegressor(n_estimators=100, random_state=1)
# flatten target variable column using ravel()
y_train_flat = y_train.squeeze().values.ravel()
fr.fit(X_train, y_train_flat)
y3 = fr.predict(X_test)
y3
```

```python
# SVR

svr_model = SVR(kernel='rbf', C=1.0, gamma='scale') 
# Reshape y_train
y_train_reshaped = np.ravel(y_train)
svr_model.fit(X_train, y_train_reshaped)
y4 = svr_model.predict(X_test)
y4
```

```python
# KNN

knn = KNeighborsRegressor(n_neighbors=3)
knn.fit(X_train, y_train)
y5 = knn.predict(X_test)
y5
```

```python
# DUMMY REGRESSOR

dummy=DummyRegressor(strategy="mean")
dummy.fit(X_train, y_train)
y6=dummy.predict(X_test)
y6
```

```python
mae_simple = mean_absolute_error(y_test, y1)
mae_lasso = mean_absolute_error(y_test, y2)
mae_random_forest = mean_absolute_error(y_test, y3)
mae_svr = mean_absolute_error(y_test, y4)
mae_knn = mean_absolute_error(y_test, y5)
mae_dummy = mean_absolute_error(y_test, y6)
mae_data = {
    'Model': ['Simple Linear Regression', 'Lasso Regression', 'Random Forest', 'SVR', 'KNN', 'Dummy Regressor'],
    'MAE': [mae_simple, mae_lasso, mae_random_forest, mae_svr, mae_knn, mae_dummy]
}

mae_table = pd.DataFrame(mae_data)
mae_table
```

## 3. Feature Selection

Use SelectKBest feature selection technique.
Reduce the number of features to 15

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
from numpy import set_printoptions
```

```python
y_train_array = np.array(y_train).ravel()

test = SelectKBest(score_func=f_regression, k=15) 
fit = test.fit(X_train, y_train_array) 

set_printoptions(precision=1)
print(fit.scores_) 
X_new = fit.transform(X_train) 
X_new[0:15]
```

```python
X_train
```

```python
X_new.shape
```

```python
fit.get_support(indices=True)
X_test_new = fit.transform(X_test)
X_test_new.shape
```

Create new regressors with default parameters with the selected 15 features:
- Linear Regression
- LASSO regression
- Random Forest
- SVR
- KNN
- Dummy regressor

Compare the MAE values in a table

```python
# Linear Regression
lr = LinearRegression()
lr.fit(X_new, y_train_array)
lr_pred = lr.predict(X_test_new)
lr_mae = mean_absolute_error(y_test, lr_pred)

# LASSO Regression
lasso = Lasso()
lasso.fit(X_new, y_train_array)
lasso_pred = lasso.predict(X_test_new)
lasso_mae = mean_absolute_error(y_test, lasso_pred)

# Random Forest
rf = RandomForestRegressor()
rf.fit(X_new, y_train_array)
rf_pred = rf.predict(X_test_new)
rf_mae = mean_absolute_error(y_test, rf_pred)

# SVR
svr = SVR()
svr.fit(X_new, y_train_array)
svr_pred = svr.predict(X_test_new)
svr_mae = mean_absolute_error(y_test, svr_pred)

# KNN
knn = KNeighborsRegressor()
knn.fit(X_new, y_train_array)
knn_pred = knn.predict(X_test_new)
knn_mae = mean_absolute_error(y_test, knn_pred)

# Dummy Regressor
dummy = DummyRegressor(strategy='mean')
dummy.fit(X_new, y_train_array)
dummy_pred = dummy.predict(X_test_new)
dummy_mae = mean_absolute_error(y_test, dummy_pred)

mae_data = {
    'Model': ['Linear Regression', 'LASSO Regression', 'Random Forest', 'SVR', 'KNN', 'Dummy Regressor'],
    'MAE': [lr_mae, lasso_mae, rf_mae, svr_mae, knn_mae, dummy_mae]
}

mae_df = pd.DataFrame(mae_data)

mae_df
```

Compare the the two tables (all features vs top15 features).
Discuss the results

The difference is very big. When using all features, the mistake is very small which is a sign of overfitting. With 15 features, the mistake is quite normal.

## 4. Hyperparameter Optimization

Use the dataset with the selected 15 features.

### 4.1 Random Search
Use Random Search CV to optimize the parameters for:
- LASSO regression
- Random Forest
- SVR

Use appropriate  n_iter, so that it ends in a reasonable time
 <br>Use CV = 3

Use Cross Validation for optimization, and the test set for final evaluation
<br> <b>DO NOT OPTIMIZE ON THE TEST SET </b>

Show the 2 comparison tables: 
- on the CV  
- on the test set

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import cross_val_score
```

```python
param_grid_lasso = {
    'alpha': np.logspace(-4, 4, 100)
}

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt', 2, 3, 5, 6, 8, 'None']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [1, 2, 5, 8, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 3, 4, 5]
# Method of selecting samples for training each tree
bootstrap = [True, False]

random_search = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}

lasso = Lasso()

lasso_random = RandomizedSearchCV(estimator=lasso, 
                                  param_distributions=param_grid_lasso, 
                                  n_iter=50, 
                                  cv=3, 
                                  verbose=2, 
                                  random_state=1, 
                                  n_jobs=-1)

lasso_random.fit(X_new, y_train_array)

print("Best parameters for LASSO Regression:", lasso_random.best_params_)

print("CV results for LASSO Regression:")
lasso_cv_results = pd.DataFrame(lasso_random.cv_results_)
print(lasso_cv_results[['mean_test_score', 'std_test_score', 'params']])

lasso_test_pred = lasso_random.predict(X_test_new)
lasso_test_mae = mean_absolute_error(y_test, lasso_test_pred)
print("MAE on test set for LASSO Regression:", lasso_test_mae)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor

param_grid_rf = {
    'n_estimators': [100, 200, 300, 400, 500],
    'max_features': ['auto', 'sqrt', 'log2', None],
    'max_depth': [10, 20, 30, 40, 50, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

rf = RandomForestRegressor()
rf_random = RandomizedSearchCV(estimator=rf, 
                               param_distributions=param_grid_rf, 
                               n_iter=50, 
                               cv=3, 
                               scoring='neg_mean_absolute_error', 
                               verbose=2, 
                               random_state=1, 
                               n_jobs=-1)

rf_random.fit(X_new, y_train_array)


print("Best parameters for Random Forest:", rf_random.best_params_)

print("CV results for Random Forest:")
rf_cv_results = pd.DataFrame(rf_random.cv_results_)
print(rf_cv_results[['mean_test_score', 'std_test_score', 'params']])

rf_test_pred = rf_random.predict(X_test_new)
rf_test_mae = mean_absolute_error(y_test, rf_test_pred)
print("MAE on test set for Random Forest:", rf_test_mae)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVR

param_grid_svr = {
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'C': [0.1, 1, 10, 100],
    'gamma': ['scale', 'auto']
}

svr = SVR()

svr_random = RandomizedSearchCV(estimator=svr, 
                                param_distributions=param_grid_svr, 
                                n_iter=50, 
                                cv=3, 
                                scoring='neg_mean_absolute_error', 
                                verbose=2, 
                                random_state=1, n_jobs=-1)

svr_random.fit(X_new, y_train_array)


print("Best parameters for SVR:", svr_random.best_params_)


print("CV results for SVR:")
svr_cv_results = pd.DataFrame(svr_random.cv_results_)
print(svr_cv_results[['mean_test_score', 'std_test_score', 'params']])


svr_test_pred = svr_random.best_estimator_.predict(X_test_new)
svr_test_mae = mean_absolute_error(y_test, svr_test_pred)
print("MAE on test set for SVR:", svr_test_mae)
```
